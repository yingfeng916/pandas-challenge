```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name"])
```

```python
school_data_complete.head()
```

```python
total_schools = len(school_data_complete["school_name"].unique())
total_schools
```

```python
total_students = len(school_data_complete)
total_students
```

```python
total_budget = school_data["budget"].sum()
total_budget
```

```python
average_math = school_data_complete["math_score"].mean()
average_math
```

```python
average_reading = school_data_complete["reading_score"].mean()
average_reading
```

```python
passing_math_df = school_data_complete.loc[school_data_complete["math_score"] >= 70]
passing_math_df
```

```python
percentage_passed_math = {len(passing_math_df["math_score"])/ total_students * 100}
percentage_passed_math
```

```python
passing_reading_df = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
passing_reading_df
```

```python
percentage_passed_reading = {len(passing_reading_df["reading_score"])/ total_students * 100}
percentage_passed_reading
```

```python
passed_both_df = school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
passed_both_df
```

```python
percentage_passed_both = {len(passed_both_df)/ total_students * 100}
percentage_passed_both
```

```python
district_data = {'Total Schools': total_schools,
      'Total Students': total_students,
      'Total Budget': total_budget,
      'Average Math Score': average_math,
      'Average Reading Score': average_reading,
      '% Passing Math': percentage_passed_math,
      '% Passing Reading': percentage_passed_reading,
      '% Overall Passing': percentage_passed_both,
        }
district_df = pd.DataFrame(district_data, columns = ['Total Schools', 'Total Students', 'Total Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing'])
```

```python
district_df
```

```python
group_students_df = (school_data_complete[["school_name", "student_name"]]).groupby(["school_name"]).count()
group_students_df
```

```python
school_df = school_data_complete[["school_name", "budget", "math_score", "reading_score"]]
school_df.head(10)
```

```python
school_group_df = school_df.groupby(["school_name"]).mean()
school_group_df
```

```python
school_group_df = pd.merge(group_students_df, school_group_df, how="left", on=["school_name"])
school_group_df
```

```python
type_df = school_data_complete[["school_name", "type"]]
type_group_df = type_df.groupby(["school_name"]).first()
type_group_df
```

```python
school_group_two_df = pd.merge(school_group_df, type_group_df, how="left", on=["school_name"])
school_group_two_df
```

```python
per_student_budget = school_group_two_df["budget"] / school_group_two_df["student_name"]
per_student_budget_df = pd.DataFrame(per_student_budget)
per_student_budget_df.columns = ['Per Student Budget']
per_student_budget_df
```

```python
school_group_three_df = pd.merge(school_group_two_df, per_student_budget_df, how="left", on=["school_name"])
school_group_three_df
```

```python
passing_math_school = passing_math_df[["school_name","math_score"]].groupby(["school_name"]).count()
passing_math_school 
```

```python
percentage_passing_math = passing_math_school["math_score"].div(group_students_df["student_name"])*100
percentage_passing_math_df = pd.DataFrame(percentage_passing_math)
percentage_passing_math_df.columns = ['% Passing Math']
percentage_passing_math_df
```

```python
passing_reading_school = passing_reading_df[["school_name","reading_score"]].groupby(["school_name"]).count()
percentage_passing_reading = passing_reading_school["reading_score"].div(group_students_df["student_name"])*100
percentage_passing_reading_df = pd.DataFrame(percentage_passing_reading)
percentage_passing_reading_df.columns = ['% Passing Reading']
percentage_passing_reading_df
```

```python
passing_both_school = passed_both_df[["school_name","reading_score"]].groupby(["school_name"]).count()
percentage_passing_both = passing_both_school["reading_score"].div(group_students_df["student_name"])*100
percentage_passing_both_df = pd.DataFrame(percentage_passing_both)
percentage_passing_both_df.columns = ['% Passing Both']
percentage_passing_both_df
```

```python
one = pd.merge(school_group_three_df, percentage_passing_math_df, how="left", on=["school_name"])
two = pd.merge(one, percentage_passing_reading_df, how="left", on=["school_name"])
school_summary_df = pd.merge(two, percentage_passing_both_df, how="left", on=["school_name"])
school_summary_df.head()
```

```python
school_summary_df = school_summary_df.rename(columns={"student_name": "Total Students", "budget": "Total School Budget", 
                                  "math_score": "Average Math Score", "reading_score": "Average Reading Score",
                                 "type": "School Type", "% Passing Both": "% Overall Passing"})
school_summary_df = school_summary_df[["School Type",
      "Total Students",
      "Total School Budget",
      "Per Student Budget",
      "Average Math Score",
      "Average Reading Score",
      "% Passing Math",
      "% Passing Reading",
      "% Overall Passing"]]

school_summary_df.head()
```

```python
sort_over_passing_df = school_summary_df.sort_values(['% Overall Passing'], ascending=False)
sort_over_passing_df.head(5)
```

```python
sort_over_passing_df = school_summary_df.sort_values(['% Overall Passing'], ascending=True)
sort_over_passing_df.head(5)
```

```python
bygrade_9 = school_data_complete[school_data_complete["grade"].str.contains("9")]
math_bygrade_9 = bygrade_9[["school_name", "math_score"]].groupby(["school_name"]).mean()
math_bygrade_9 = math_bygrade_9.rename(columns={"math_score": "9th"})
math_bygrade_9
```

```python
bygrade_10 = school_data_complete[school_data_complete["grade"].str.contains("10")]
math_bygrade_10 = bygrade_10[["school_name", "math_score"]].groupby(["school_name"]).mean()
math_bygrade_10 = math_bygrade_10.rename(columns={"math_score": "10th"})
math_bygrade_10
```

```python
bygrade_11 = school_data_complete[school_data_complete["grade"].str.contains("11")]
math_bygrade_11 = bygrade_11[["school_name", "math_score"]].groupby(["school_name"]).mean()
math_bygrade_11 = math_bygrade_11.rename(columns={"math_score": "11th"})
math_bygrade_11
```

```python
bygrade_12 = school_data_complete[school_data_complete["grade"].str.contains("12")]
math_bygrade_12 = bygrade_12[["school_name", "math_score"]].groupby(["school_name"]).mean()
math_bygrade_12 = math_bygrade_12.rename(columns={"math_score": "12th"})
math_bygrade_12
```

```python
one = pd.merge(math_bygrade_9, math_bygrade_10, how="left", on=["school_name"])
two = pd.merge(one, math_bygrade_11, how="left", on=["school_name"])
math_bygrade_summary_df = pd.merge(two, math_bygrade_12, how="left", on=["school_name"])
math_bygrade_summary_df
```

```python
reading_bygrade_9 = bygrade_9[["school_name", "reading_score"]].groupby(["school_name"]).mean()
reading_bygrade_9 = reading_bygrade_9.rename(columns={"reading_score": "9th"})

reading_bygrade_10 = bygrade_10[["school_name", "reading_score"]].groupby(["school_name"]).mean()
reading_bygrade_10 = reading_bygrade_10.rename(columns={"reading_score": "10th"})

reading_bygrade_11 = bygrade_11[["school_name", "reading_score"]].groupby(["school_name"]).mean()
reading_bygrade_11 = reading_bygrade_11.rename(columns={"reading_score": "11th"})

reading_bygrade_12 = bygrade_12[["school_name", "reading_score"]].groupby(["school_name"]).mean()
reading_bygrade_12 = reading_bygrade_12.rename(columns={"reading_score": "12th"})

one = pd.merge(reading_bygrade_9, reading_bygrade_10, how="left", on=["school_name"])
two = pd.merge(one, reading_bygrade_11, how="left", on=["school_name"])
reading_bygrade_summary_df = pd.merge(two, reading_bygrade_12, how="left", on=["school_name"])
reading_bygrade_summary_df
```

```python
bins = [0, 585, 630, 645, 676]

group_names = ["<584", "$585-629", "$630-644", "$645-675"]

school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels=group_names, include_lowest=True)

math_school_spending_df = ((school_summary_df[["Spending Ranges (Per Student)", "Average Math Score"]]).groupby(["Spending Ranges (Per Student)"]).mean()).round(2)
math_school_spending_df
```

```python
reading_school_spending_df = ((school_summary_df[["Spending Ranges (Per Student)", "Average Reading Score"]]).groupby(["Spending Ranges (Per Student)"]).mean()).round(2)
reading_school_spending_df
```

```python
percentage_math_school_spending_df = ((school_summary_df[["Spending Ranges (Per Student)", "% Passing Math"]]).groupby(["Spending Ranges (Per Student)"]).mean()).round(2)
percentage_reading_school_spending_df = ((school_summary_df[["Spending Ranges (Per Student)", "% Passing Reading"]]).groupby(["Spending Ranges (Per Student)"]).mean()).round(2)
percentage_both_school_spending_df = ((school_summary_df[["Spending Ranges (Per Student)", "% Overall Passing"]]).groupby(["Spending Ranges (Per Student)"]).mean()).round(2)

```

```python
one = pd.merge(math_school_spending_df, reading_school_spending_df, how="left", on=["Spending Ranges (Per Student)"])
two = pd.merge(one, percentage_math_school_spending_df, how="left", on=["Spending Ranges (Per Student)"])
three = pd.merge(two, percentage_reading_school_spending_df, how="left", on=["Spending Ranges (Per Student)"])
scores_school_spending_df = pd.merge(three, percentage_both_school_spending_df, how="left", on=["Spending Ranges (Per Student)"])
scores_school_spending_df
```

```python
bins_2 = [0, 1000, 2000, 5000]

group_names_2 = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], bins_2, labels=group_names_2, include_lowest=True)

math_school_size_df = (school_summary_df[["School Size", "Average Math Score"]]).groupby(["School Size"]).mean()
math_school_size_df
```

```python
reading_school_size_df = (school_summary_df[["School Size", "Average Reading Score"]]).groupby(["School Size"]).mean()
percentage_math_school_size_df = (school_summary_df[["School Size", "% Passing Math"]]).groupby(["School Size"]).mean()
percentage_reading_school_size_df = (school_summary_df[["School Size", "% Passing Reading"]]).groupby(["School Size"]).mean()
percentage_both_school_size_df = (school_summary_df[["School Size", "% Overall Passing"]]).groupby(["School Size"]).mean()

```

```python
one = pd.merge(math_school_size_df, reading_school_size_df, how="left", on=["School Size"])
two = pd.merge(one, percentage_math_school_size_df, how="left", on=["School Size"])
three = pd.merge(two, percentage_reading_school_size_df, how="left", on=["School Size"])
scores_school_size_df = pd.merge(three, percentage_both_school_size_df, how="left", on=["School Size"])
scores_school_size_df
```

```python
scores_type = sort_over_passing_df.groupby(["School Type"]).mean()
scores_type_df = scores_type[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
scores_type_df
```
